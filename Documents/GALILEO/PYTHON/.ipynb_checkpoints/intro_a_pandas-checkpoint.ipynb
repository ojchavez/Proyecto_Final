## Merge, Join, Concatenate

```python
df1 = pd.DataFrame({'A': ['A1', 'A2', 'A3', 'A4'],
                    'B': ['B1', 'B2', 'B3', 'B4'],
                    'C': ['C1', 'C2', 'C3', 'C4'],
                    'D': ['D1', 'D2', 'D3', 'D4']}, index=[0,1,2,3])
df1
```

```python
df2 = pd.DataFrame({'A': ['A5', 'A6', 'A7', 'A8'],
                    'B': ['B5', 'B6', 'B7', 'B8'],
                    'C': ['C5', 'C6', 'C7', 'C8'],
                    'D': ['D5', 'D6', 'D7', 'D8']}, index=[0,1,2,3])
df2
```

```python
df3 = pd.DataFrame({'A': ['A9', 'A10', 'A11', 'A12'],
                    'B': ['B9', 'B10', 'B11', 'B12'],
                    'C': ['C9', 'C10', 'C11', 'C12'],
                    'D': ['D9', 'D10', 'D11', 'D12']}, index=[0,1,2,3])
df3
```

```python
dfConcat = pd.concat([df1, df2, df3], ignore_index=True)
dfConcat
```

```python
dfConcat = pd.concat([df1, df2, df3], keys=['a', 'b', 'c'])
dfConcat
```

```python
dfConcat.loc['b']
```

```python
df4 = pd.DataFrame({'A': ['A9', 'A10', 'A11', 'A12'],
                    'B': ['B9', 'B10', 'B11', 'B12'],
                    'C': ['C9', 'C10', 'C11', 'C12'],
                    'D': ['D9', 'D10', 'D11', 'D12']}, index=[7,8,9,10])
df4
```

```python
dfConcat = pd.concat([df1, df4], axis=1)
dfConcat
```

### Merge

```python
left = pd.DataFrame({'Key': ['K0', 'K1', 'K2', 'K3'],
                    'H': ['H9', 'H10', 'H11', 'H12'],
                    'C': ['C9', 'C10', 'C11', 'C12'],
                    'D': ['D9', 'D10', 'D11', 'D12']}, index=[0,1,2,3])
left
```

```python
right = pd.DataFrame({'Key': ['K0', 'K1', 'K2', 'K3'],
                    'B': ['B9', 'B10', 'B11', 'B12'],
                    'M': ['M9', 'M10', 'M11', 'M12']}, index=[0,1,2,3])
right
```

```python
pd.merge(left, right, on='Key')
```

### Join

```python
A = pd.DataFrame({'B': ['B9', 'B10', 'B11', 'B12'], 'M': ['M9', 'M10', 'M11', 'M12']}, index=[0,1,2,3])
B = pd.DataFrame({'H': ['H9', 'H10', 'H11', 'H12'], 'C': ['C9', 'C10', 'C11', 'C12'], 'D': ['D9', 'D10', 'D11', 'D12']}, index=[0,1,2,4])
```

```python
A.join(B)
```

```python
A = pd.DataFrame({'A': ['A0', 'A1', 'A2'], 'B': ['B0', 'B1', 'B2'],
                     'C': ['K0', 'K1', 'K2']})
A
```

```python
B = pd.DataFrame({'D': ['C0', 'C1', 'C2'], 'E': ['E0', 'E1', 'E2'],
                     'F': ['K0', 'K1', 'K3']})
B
```

```python
result = A.join(B.set_index(['F']), on=['C'], how='inner')
result
```

```python
result = A.join(B.set_index(['F']), on=['C'], how='left')
result
```

```python
result = left.join(right)
```
